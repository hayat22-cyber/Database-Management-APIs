```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("weather_pipeline.db")
```

```python
query = "SELECT * FROM house_prices LIMIT 5"
pd.read_sql(query, conn)
```

```python
query = "SELECT COUNT(*) AS total_properties FROM house_prices"
pd.read_sql(query, conn)
```

```python
query = "SELECT AVG(Price) AS average_price FROM house_prices"
pd.read_sql(query, conn)
```

```python
query = """
SELECT Location, AVG(Price) AS avg_price
FROM house_prices
GROUP BY Location
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT Property_ID, Location, Property_Type, Price
FROM house_prices
ORDER BY Price DESC
LIMIT 1
"""
pd.read_sql(query, conn)
```

```python
conn.close()
print("Database connection closed")
```
